The uneven distribution of information and communication technologies in society.

# Data Analysis

## State Overview

To start, I decided to group the data by the state, so we can take a look at some trends. The shape of this dataframe consists of 51 rows and that's because it contains DC as a state. Also, when aggregating as the mean function is doing, it accounts for null values and does not include them in the calculation, which is why I haven't had to drop the null values in cleaning. 

```python
overview = df.groupby(['state'],as_index = False).mean()
overview
```

```python
overview.shape

#considers DC a state 
```

```python
dd1 = overview[['state','poverty','no_comp','no_internet']]
dd1
```

```python
import seaborn as sns

#plt.scatter(overview["poverty"], overview["no_internet"])
sns.set_style("white")
gridobj = sns.lmplot(x="poverty", y="no_internet", data= overview, 
                     height=5, aspect=1.3, robust=True, palette='tab10', 
                     scatter_kws=dict(s=60, linewidths=.7, edgecolors='black'))
```

```python
#plt.scatter(overview["poverty"], overview["no_comp"])

sns.set_style("white")
gridobj = sns.lmplot(x="poverty", y="no_comp", data= overview, 
                     height=5, aspect=1.3, robust=True, palette='tab10', 
                     scatter_kws=dict(s=60, linewidths=.7, edgecolors='black'))
```

## Top 25% of states with the lowest poverty rate 

```python
lowest_poverty = dd1.sort_values(by=['poverty'],ascending=True, kind='quicksort')[['state','poverty']].head(13)
lowest_poverty
```

```python
#Try to make a lollipop chart with the rate above the circle 

#fig, axes = plt.subplots()
#axes.vlines(lowest_poverty, ymin=0, ymax=lowest_poverty)
#axes.plot(lowest_poverty, lowest_poverty, "o")
#axes.set_ylim(0)

import pandas as pd
import matplotlib.pyplot as plt

my_range=range(0,len(lowest_poverty))

plt.figure(figsize =(10,6),tight_layout=True)
plt.margins(.1)
plt.stem(lowest_poverty['poverty'])
plt.xticks(my_range,lowest_poverty['state'],rotation = 90)
plt.xlabel("State")
plt.ylabel("%Poverty Rate")
plt.title("Top 25% of States w/ Lowest Poverty Rate")

for idx, row in enumerate(lowest_poverty.itertuples()):
    plt.text(idx, row.poverty+.2, s=f'{row.poverty/100:.2%}', horizontalalignment= 'center', verticalalignment='bottom', fontsize=10)

plt.show()
```

## Correlogram 

A graph to show the correlation between variables.

```python
corr = overview[['state','poverty','no_comp','no_internet']].corr()
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='dark')
%matplotlib inline
plt.figure(figsize=(12,8))
plt.title('Correlation Between Poverty Rate and Lack of Access to Internet & Computers by State')
sns.heatmap(corr, annot=True)
```

```python
#Rank the poverty level 

no_internet = overview.sort_values(by=['no_internet'],ascending=False, kind='quicksort')[['state','poverty','no_internet']]
no_internet

#Creating a column called rank which will rank the states in descending order with respect to poverty level 
no_internet["Rank"] = no_internet["poverty"].rank(ascending=False)


no_internet

#https://www.dataindependent.com/pandas/pandas-rank/
```

```python
no_internet = dd1.sort_values(by=['no_internet'],ascending=False, kind='quicksort')[['state','poverty','no_internet']].head(13)
no_internet
```

## Combination Chart

```python
#Make a combination chart

fig = plt.figure(figsize =(16,5))
ax1 = plt.subplot(1,1,1)
ax1.bar(no_internet.state, no_internet.no_internet)
plt.xlabel('State')
plt.ylabel('%No Internet')

#ax1.set_xlim(0,no_internet.no_internet.max()*1.25)

ax2= ax1.twinx()
ax2.plot(no_internet.state, no_internet.poverty,'o-',color='red', linewidth = 2)
plt.ylabel('%Poverty Rate')
plt.title('No Internet & Poverty Rate by State')


plt.show()


#Resource 

#https://www.youtube.com/watch?v=PTy7hw7goeA

#Need to convert to percentages!!
```
