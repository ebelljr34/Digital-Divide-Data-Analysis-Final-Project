# What is the digital divide?

The digital divide, as defined by Britannica, is the uneven distribution of information and communication technologies in society. As the world of technology is rapidly increasing, so is the gap for individuals who lack access to these same worldchanging technologies. This issue becomes increasingly important when we begin to look critically at the individuals and commuities this technological inequity affects the most.
